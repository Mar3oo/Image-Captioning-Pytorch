## Step 1: Initialize the Flickr8k Dataset

We begin by initializing the [Flickr8k dataset](https://github.com/jbrownlee/Datasets/blob/master/Flickr8k_Dataset.names) that you will use to obtain the data.


```python
import os
import pandas as pd

# Define the path to the local Flickr8k dataset
dataDir = './flickr8k'
images_dir = os.path.join(dataDir, 'Images')  # folder containing images
captions_file = os.path.join(dataDir, 'captions.txt')  # file containing captions

# Load the captions data into a DataFrame
captions_data = pd.read_csv(captions_file)

# Prepare a dictionary to store the image captions
image_captions = {}

# Populate the image_captions dictionary
for index, row in captions_data.iterrows():
    image_id = row['image']
    caption_text = row['caption']
    if image_id not in image_captions:
        image_captions[image_id] = []
    image_captions[image_id].append(caption_text)

# Get all the image IDs (corresponding to image files in the dataset)
image_ids = [image_id for image_id in os.listdir(images_dir) if image_id.endswith('.jpg')]

# Check if the images and captions are properly loaded
print(f"Loaded {len(image_captions)} images with captions.")
print(f"Sample caption for image '1000268201_693b08cb0e.jpg': {image_captions.get('1000268201_693b08cb0e.jpg', [])[:1]}")
```

## Step 2: Plot a Sample Image

Next, we plot a random image from the dataset, along with its five corresponding captions.  Each time you run the code cell below, a different image is selected.  

In the project, you will use this dataset to train your own model to generate captions from images!

```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import random

# Pick a random image and obtain the corresponding image path
image_id = random.choice(list(image_captions.keys()))  # Randomly select an image ID
image_path = os.path.join(images_dir, image_id)  # Full path to the image

# Print the image path and visualize the corresponding image
print(image_path)
I = io.imread(image_path)  # Read the image
plt.axis('off')
plt.imshow(I)
plt.show()

# Load and display captions for the selected image
captions = image_captions.get(image_id, [])
print(f"Captions for {image_id}:")
for caption in captions:
    print(caption)
```
